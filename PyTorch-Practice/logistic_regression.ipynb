```python
import torch.nn.functional as F
import torch.optim as optim
from torch import tensor
from torch import nn
from torch import sigmoid

x_data = tensor([[1.0],[2.0],[3.0],[4.0]])
y_data = tensor([[0.],[0.],[1.],[1.]])
```

```python
# Use sigmoid
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = nn.Linear(1,1)
    
    def forward(self, x):
        y_pred = sigmoid(self.linear(x))
        return y_pred
```

```python
model = Model()
criterion = nn.BCELoss(reduction = 'mean')
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# Training loop
for epoch in range(1000):
    y_pred = model(x_data)
    
    loss = criterion(y_pred, y_data)
    print(f"Epoch {epoch + 1}/1000 | Loss : {loss.item():.4f}")
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
# After training
print("\nLet's predict the hours need to score above 50%")
print("=" * 50)
hour_var = model(tensor([[1.0]]))
print(f'Prediction after 1 hour of training : {hour_var.item():.4f}| Above 50% : {hour_var.item() > 0.5}')
hour_var = model(tensor([[7.0]]))
print(f'Prediction after 7 hours of training: {hour_var.item():.4f} | Above 50%: { hour_var.item() > 0.5}')
```

```python
from torch import nn, optim, from_numpy
import numpy as np

xy = np.loadtxt('./diabetes.csv', delimiter = ',', dtype = np.float32)
x_data = from_numpy(xy[:,0:-1])
y_data = from_numpy(xy[:,[-1]])

print(f'X\'s shape : {x_data.shape} | Y\'s shape : {y_data.shape}')
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.l1 = nn.Linear(8,6)
        self.l2 = nn.Linear(6,4)
        self.l3 = nn.Linear(4,1)
        
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        out1 = self.sigmoid(self.l1(x))
        out2 = self.sigmoid(self.l2(out1))
        y_pred = self.sigmoid(self.l3(out2))
        return y_pred
    
```

```python
# SGD
model = Model()
criterion = nn.BCELoss(reduction = 'mean')
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(100):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    print(f'Epoch : {epoch + 1}/100 | Loss : {loss.item():.4f}')
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# Adam
model = Model()
criterion = nn.BCELoss(reduction = 'mean')
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
for epoch in range(100):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    print(f'Epoch : {epoch + 1}/100 | Loss : {loss.item():.4f}')
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
