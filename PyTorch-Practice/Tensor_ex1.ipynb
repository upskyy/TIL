```python
import torch
t = torch.FloatTensor([0,1,2,3,4,5,6])
print(t)
```

```python
print(t.dim())
print(t.shape)
print(t.size())
print(t[0],t[2],t[-1])
print(t[2:4],t[3:-1])
print(t[:2],t[4:])
```

```python
m1 = torch.FloatTensor([3,3])
m2 = torch.FloatTensor([2,2])
print(m1 + m2)
```

```python
m1 = torch.FloatTensor([1,2])
m2 = torch.FloatTensor([3])
print(m1 + m2)
```

```python
m1 = torch.FloatTensor([1,2])
m2 = torch.FloatTensor([[3],[4]])
print(m1 + m2)
```

```python
print("Mul vs Matmul")
m1 = torch.FloatTensor([[1,2],[3,4]])
m2 = torch.FloatTensor([[1],[2]])
print("Shape of matrix 1:",m1.shape)
print("Shape of matrix 2: {}".format(m2.shape))
print(m1.matmul(m2))  # 2 x 1

print(m1 * m2)  # 2 x 2
print(m1.mul(m2))  # 2 x 2
```

```python
t = torch.FloatTensor([1,2])
print(t.mean())
```

```python
t = torch.IntTensor([1,2])  # Can't use mean() on integers
try:
    print(t.mean())
except Exception as exc:
    print(exc)
```

```python
t = torch.FloatTensor([[1,2],[3,4]])
print(t)
```

```python
print(t.mean())
print(t.mean(dim = 0))
print(t.mean(dim = 1))
print(t.mean(dim = -1))
```

```python
print(t.sum())
print(t.sum(dim = 0))
print(t.sum(dim = 1))
print(t.sum(dim = -1))
```

```python
print(t.max())
```

```python
print(t.max(dim = 0))   # Return two values : max and argmax
print("Max :",t.max(dim = 0)[0])
print("Argmax : {}".format(t.max(dim = 0)[1]))
```

```python
print(t.max(dim = 1))
print(t.max(dim = -1))
```
